```python
#1.Crie um dicionário chamado qtd_produtos que armazene a quantidade de produtos comprados em um e-commerce
qtd_produtos = {
    'João': 3,
    'Maria': 2,
    'Pedro': 4
}

#2.Imprima o dicionário completo e o tipo da variável. Lembre-se da instrução type;
print('Dicionário:', qtd_produtos)
print(type(qtd_produtos))

#3.Peça ao usuário para digitar o nome de um cliente e, se o cliente existir no dicionário,
#imprima a quantidade de produtos comprados por ele.
#Veja a lista de métodos disponíveis para dicionários para atingir esse objetivo;

cliente = input('Digite o nome de um cliente: ').capitalize().strip()
if cliente in qtd_produtos:
  print(f"Quantidade de produtos comprados por {cliente} é:", qtd_produtos[cliente])
else:
  print("Esse cliente não existe no dicionário.")

#4.Altere o valor da chave de um dos clientes para simular que ele comprou mais 5 produtos;
qtd_produtos[cliente] += 5
print(f'Compra de {cliente} atualizada:', qtd_produtos[cliente])

#5.Adicione um novo cliente ao dicionário com a quantidade inicial de 0 produtos;
qtd_produtos.update({'Bruno': 0})
print('Dicionário atualizado:', qtd_produtos)

#6.Remova um cliente do dicionário usando o método apropriado;
qtd_produtos.pop(cliente)
print(f'{cliente} removido/a do dicionário:', qtd_produtos)

#7.Imprima as chaves, os valores e os pares chave-valor do dicionário usando os métodos keys(), values() e items().
print('Keys do dicionário:', qtd_produtos.keys())
print('Valores do dicionário:', qtd_produtos.values())
print('Itens do dicionário:', qtd_produtos.items())
```
