```python
import os
import pickle
from datetime import datetime

ARQUIVO = 'teste.pkl'
lista = [1, 2, 3, 4]

def validar_data(data_str):
  try:
    datetime.strptime(data_str, "%Y-%m-%d")
    return True
  except ValueError:
    return False

def salvar_lista(lista):
  with open(ARQUIVO, 'wb') as f:
    pickle.dump(lista, f)

def carregar_lista():
  if os.path.exists(ARQUIVO):
    try:
      with open(ARQUIVO, 'rb') as f:
        return pickle.load(f)
    except:
      print("Arquivo corrompido. Criando novo.")
      return []
  return []

def validar_valor(valor_str):
  try:
    valor= float(valor_str.replace(',','.'))
    print(valor)
    return valor
  except ValueError:
    return False

input_data = input("Digite uma data no formato YYYY-MM-DD: ").strip()
while not validar_data(input_data):
   print("Data inválida! Tente novamente.")
   input_data = input("Digite uma data no formato YYYY-MM-DD: ").strip()
print('data válida')

salvar_lista(lista)
lista_carregada = carregar_lista()

print(lista_carregada)

input_valor = input("Digite um valor: ").strip()
while not validar_valor(input_valor):
   print("Valor inválido! Tente novamente.")
   input_valor = input("Digite um valor: ").strip()
```
