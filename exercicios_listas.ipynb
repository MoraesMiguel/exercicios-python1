```python
#Crie uma lista chamada ‘clientes’ contendo pelo menos 3 nomes de clientes do seu e-commerce.
clientes = ['João', 'Pedro', 'Alberto']

#1.Imprima a lista e o tipo da variável;
print('Lista de clientes:', clientes)
print('Tipo da variável:', type(clientes))

#2.Acesse e imprima o primeiro e o último elemento usando indexação;
print('Primeiro elemento:', clientes[0])
print('Último elemento:', clientes[-1])

#3.Substitua o nome do segundo cliente por outro nome e imprima a lista atualizada;
clientes_modificado = clientes.copy()
clientes_modificado[1] = 'Maria'
print('Lista modificada:', clientes_modificado)

#4.Use slicing para criar uma sublista com os dois primeiros clientes e a imprima;
print('Sublista com os dois primeiros clientes:', clientes_modificado[0:2])

#5.Adicione um novo cliente ao final da lista usando o método apropriado;
clientes_modificado.append('Lucas')
print('Adicionado Lucas no final', clientes_modificado)

#6.Remova o primeiro cliente usando o método apropriado;
clientes_modificado.pop(0)
print('Removido o 1° cliente:', clientes_modificado)

#7.Ordene a lista de clientes em ordem alfabética e imprima;
clientes_modificado.sort()
print('Lista ordenada:', clientes_modificado)

#8.Inverta a ordem da lista usando um método;
clientes_modificado.reverse()
print('Lista ordenada invertida:', clientes_modificado)

#9.Imprima o tamanho da lista usando uma função nativa.
print('Tamanho da lista:', len(clientes_modificado))
```
